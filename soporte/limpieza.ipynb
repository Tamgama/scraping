```python
df = pd.concat([alqui, ventas], axis=0, ignore_index=True)
df
```

```python
df = df.drop(['Fianza', 'Comunidad'], axis=1)

df.columns = df.columns.str.lower().str.replace(' ', '_')

df.rename(columns={'id_inmueble':'id',
                          'título':'titulo',
                          'área':'area',
                          'características':'caracteristicas',
                          'teléfono': 'tlf',
                          'última_actualización': 'ultima_actualizacion'}, inplace=True)

df
```

```python
cols = ['calle', 'barrio', 'distrito', 'ciudad']
for col in cols:
    df[col] = df[col].str.replace('\n', '', regex=False)

df['titulo'] = df['titulo'].str.replace('Alquiler de ', '', regex=False)
df['area'] = df['area'].str.replace('Área de ', '', regex=False)
df['barrio'] = df['barrio'].str.replace('Barrio ', '', regex=False)
df['distrito'] = df['distrito'].str.replace('Distrito ', '', regex=False)

# Convertir todos los valores a números, coercionando errores a NaN
df['tlf'] = pd.to_numeric(df['tlf'], errors='coerce')

# Rellenar NaN con 0 (u otro valor) y convertir a int
df['tlf'] = df['tlf'].fillna(0).astype(int)
```

```python
columnas = [
    'superficie',
    'superficie_util',
    'habitaciones',
    'baños',
    'terraza',
    'garaje',
    'estado',
    'armarios',
    'trastero',
    'orientacion',
    'amueblado',
    'calefaccion',
    'planta',
    'ascensor',
    'construccion',
    'movilidad_reducida'
]

for col in columnas:
    df[col] = None


df
```

```python
def extraer_y_asignar_caracteristicas(caracteristicas):
    # Convertir la lista en un solo string para facilitar el análisis
    caracteristicas_str = ' '.join(caracteristicas)
    
    # Extraer y asignar cada característica usando expresiones regulares
    
    # Superficie construida
    match = re.search(r'(\d+ m² construidos)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['superficie'] = match.group(0)
    else:
        df['superficie'] = None
    
    # Superficie útil
    match = re.search(r'(\d+ m² útiles)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['superficie_util'] = match.group(0)
    else:
        df['superficie_util'] = None
    
    # Habitaciones
    match = re.search(r'(\d+ habitaciones)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['habitaciones'] = match.group(0)
    else:
        df['habitaciones'] = None
    
    # Baños
    match = re.search(r'(\d+ baños)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['baños'] = match.group(0)
    else:
        df['baños'] = None
    
    # Planta
    match = re.search(r'Planta (\d+)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['planta'] = match.group(1)  # Solo el número de planta
    else:
        df['planta'] = None
    
    # Ascensor
    df['ascensor'] = 'Con ascensor' if re.search(r'Con ascensor', caracteristicas_str, re.IGNORECASE) else None
    
    # Terraza
    df['terraza'] = 'Terraza' if re.search(r'Terraza', caracteristicas_str, re.IGNORECASE) else None
    
    # Garaje
    df['garaje'] = 'Plaza de garaje incluida en el precio' if re.search(r'Plaza de garaje', caracteristicas_str, re.IGNORECASE) else None
    
    # Estado
    df['estado'] = 'Segunda mano/buen estado' if re.search(r'Segunda mano/buen estado', caracteristicas_str, re.IGNORECASE) else None
    
    # Armarios
    df['armarios'] = 'Armarios empotrados' if re.search(r'Armarios empotrados', caracteristicas_str, re.IGNORECASE) else None
    
    # Orientación
    match = re.search(r'Orientación ([^,]+, [^,]+)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['orientacion'] = match.group(1)
    else:
        df['orientacion'] = None
    
    # Amueblado
    df['amueblado'] = 'Amueblado y cocina equipada' if re.search(r'Amueblado y cocina equipada', caracteristicas_str, re.IGNORECASE) else None
    
    # Calefacción
    df['calefaccion'] = 'Calefacción central' if re.search(r'Calefacción central', caracteristicas_str, re.IGNORECASE) else None
    
    # Trastero
    df['trastero'] = 'Trastero' if re.search(r'Trastero', caracteristicas_str, re.IGNORECASE) else None
    


df.apply(lambda row: extraer_y_asignar_caracteristicas(row['caracteristicas']), axis=1)


df
```

```python
# Patrones de expresión regular
patterns = {
    'superficie': re.compile(r'(\d+)\s*m² construidos'),
    'superficie_util': re.compile(r'(\d+)\s*m² útiles'),
    'habitaciones': re.compile(r'(\d+)\s*habitaciones?'),
    'baños': re.compile(r'(\d+)\s*baños?'),
    'terraza': re.compile(r'Terraza|Balcón'),
    'garaje': re.compile(r'Plaza de garaje'),
    'estado': re.compile(r'Segunda mano/buen estado'),
    'armarios': re.compile(r'Armarios empotrados'),
    'trastero': re.compile(r'Trastero'),
    'orientacion': re.compile(r'Orientación\s([\w\s,]+)'),
    'amueblado': re.compile(r'(Amueblado y cocina equipada|Cocina sin equipar y casa sin amueblar)'),
    'calefaccion': re.compile(r'Calefacción (central|individual|No disponible calefaccion)'),
    'planta': re.compile(r'Planta\s(\d+)(ª|\s)(exterior|interior)'),
    'ascensor': re.compile(r'(Con ascensor|Sin ascensor)'),
    'construccion': re.compile(r'Construido en (\d{4})'),
    'movilidad_reducida': re.compile(r'Solo acceso exterior adaptado para personas con movilidad reducida')
}

def extract_data(row, patterns):
    # Inicializa el diccionario con valores nulos para todas las columnas
    data = {col: np.nan for col in columnas}
    
    for item in row:
        for key, pattern in patterns.items():
            match = pattern.search(item)
            if match:
                if key == 'planta':
                    data['planta'] = match.group(1)
                    data['exterior_interior'] = match.group(3)
                elif key == 'ascensor':
                    data[key] = 'Sí' if 'Con' in match.group(0) else 'No'
                elif key == 'amueblado':
                    data[key] = 'Amueblado' if 'Amueblado' in match.group(0) else 'No amueblado'
                elif key == 'calefaccion':
                    calef = match.group(1)
                    data[key] = calef if calef != 'No disponible calefaccion' else 'No disponible'
                elif key == 'orientacion':
                    data[key] = match.group(1)
                else:
                    data[key] = match.group(1)
                break  # Detener la búsqueda una vez encontrado un patrón
    return data

# Aplicar la función a la columna 'caracteristicas'
df_extracted = df['caracteristicas'].apply(lambda row: extract_data(row, patterns))

# Convertir la serie de diccionarios a un DataFrame
df_extracted = pd.DataFrame(list(df_extracted))

df_extracted
```
