```python
inmuebles = pd.concat([df, dff], axis=0, ignore_index=True)
inmuebles
```

```python
inmuebles = inmuebles.drop(['Fianza', 'Comunidad'], axis=1)

inmuebles.columns = inmuebles.columns.str.lower().str.replace(' ', '_')

inmuebles.rename(columns={'id_inmueble':'id',
                          'título':'titulo',
                          'área':'area',
                          'características':'caracteristicas',
                          'teléfono': 'tlf',
                          'última_actualización': 'ultima_actualizacion'}, inplace=True)

inmuebles
```

```python
cols = ['calle', 'barrio', 'distrito', 'ciudad']
for col in cols:
    inmuebles[col] = inmuebles[col].str.replace('\n', '', regex=False)

inmuebles['titulo'] = inmuebles['titulo'].str.replace('Alquiler de ', '', regex=False)
inmuebles['area'] = inmuebles['area'].str.replace('Área de ', '', regex=False)
inmuebles['barrio'] = inmuebles['barrio'].str.replace('Barrio ', '', regex=False)
inmuebles['distrito'] = inmuebles['distrito'].str.replace('Distrito ', '', regex=False)

# Convertir todos los valores a números, coercionando errores a NaN
inmuebles['tlf'] = pd.to_numeric(inmuebles['tlf'], errors='coerce')

# Rellenar NaN con 0 (u otro valor) y convertir a int
inmuebles['tlf'] = inmuebles['tlf'].fillna(0).astype(int)
```

```python
# Patrones regex actualizados
patterns = {
    'superficie_construida': re.compile(r'(\d+)\s*m² construidos'),
    'superficie_util': re.compile(r'(\d+)\s*m² útiles'),
    'habitaciones': re.compile(r'(\d+)\s*habitaciones?'),
    'baños': re.compile(r'(\d+)\s*baños?'),
    'terraza': re.compile(r'Terraza|Balcón'),
    'garaje': re.compile(r'Plaza de garaje'),
    'estado': re.compile(r'Segunda mano/buen estado'),
    'armarios': re.compile(r'Armarios empotrados'),
    'trastero': re.compile(r'Trastero'),
    'orientacion': re.compile(r'Orientación\s([\w\s,]+)'),
    'amueblado': re.compile(r'(Amueblado y cocina equipada|Cocina sin equipar y casa sin amueblar)'),
    'calefaccion': re.compile(r'Calefacción (central|individual|No disponible calefaccion)'),
    'planta': re.compile(r'Planta\s(\d+)(ª|\s)(exterior|interior)'),
    'ascensor': re.compile(r'(Con ascensor|Sin ascensor)'),
    'anio_construccion': re.compile(r'Construido en (\d{4})'),
    'movilidad_reducida': re.compile(r'Solo acceso exterior adaptado para personas con movilidad reducida')
}

def extract_data(row, patterns):
    data = {}
    for item in row:
        matched = False
        for key, pattern in patterns.items():
            match = pattern.search(item)
            if match:
                if key == 'planta':
                    data['planta'] = match.group(1)
                    data['exterior_interior'] = match.group(3)
                elif key == 'ascensor':
                    data[key] = 'Sí' if 'Con' in match.group(0) else 'No'
                elif key == 'amueblado':
                    data[key] = 'Amueblado' if 'Amueblado' in match.group(0) else 'No amueblado'
                elif key == 'calefaccion':
                    calef = match.group(1)
                    data[key] = calef if calef != 'No disponible calefaccion' else 'No disponible'
                else:
                    data[key] = match.group(1) if len(match.groups()) > 0 else True
                matched = True
        if not matched:
            # Si el item no coincide con ningún patrón, lo añadimos con NaN
            data[item] = np.nan
    return pd.Series(data)

# # Ejemplo de DataFrame con las listas en la columna 'descripcion'
# df = pd.DataFrame({
#     'descripcion': [
#         ['150 m² construidos, 140 m² útiles', '4 habitaciones', '2 baños', 'Balcón', 'Plaza de garaje incluida en el precio', 
#          'Segunda mano/buen estado', 'Armarios empotrados', 'Trastero', 'Orientación este', 'Amueblado y cocina equipada', 
#          'Planta 3ª interior', 'Con ascensor', 'Construido en 2020'],
#         ['110 m² construidos', '4 habitaciones', '2 baños', 'Segunda mano/buen estado', 'Calefacción No disponible calefaccion'],
#         ['70 m² construidos', '3 habitaciones', '1 baño', 'Balcón', 'Segunda mano/buen estado', 'Amueblado y cocina equipada', 
#          'Planta 3ª exterior', 'Sin ascensor', 'Orientación suroeste'],
#         ['160 m² construidos, 150 m² útiles', '5 habitaciones', '3 baños', 'Balcón', 'Segunda mano/buen estado', 
#          'Armarios empotrados', 'Orientación norte', 'Cocina sin equipar y casa sin amueblar', 'Planta 2ª exterior', 
#          'Con ascensor', 'Solo acceso exterior adaptado para personas con movilidad reducida']
#     ]
# })

# Aplicar la función a la columna 'descripcion'
df_extracted = inmuebles['caracteristicas'].apply(lambda row: extract_data(row, patterns))

# Combinar las nuevas columnas extraídas con el DataFrame original
df_final = pd.concat([inmuebles, df_extracted], axis=1)

# Mostrar el DataFrame resultante
print(df_final)
```
