```python
import pandas as pd
import numpy as np
import re

pd.set_option("display.max_columns", None)
pd.set_option('display.max_colwidth', None)
```

```python
alquiler = pd.read_csv("../../src/alquileres-corregido.csv")
alquiler.head(1)
```

```python
alquiler = alquiler.drop(['ultima_atualizacion', 'm_construidos', 'm_utiles', 'url', 'referencia', 'ultima_actualizacion'], axis=1)
alquiler = alquiler.rename(columns={"URL" : "url", "anunciante": "perfil", 'baños': 'banos'})

cols_to_clean = alquiler[['habitaciones', 'banos', 'precio', 'precio_metro']]
for col in cols_to_clean:
    alquiler[col] = alquiler[col].str.replace('baños', '', regex=False)
    alquiler[col] = alquiler[col].str.replace('baño', '', regex=False)
    alquiler[col] = alquiler[col].str.replace('habitaciones', '', regex=False)
    alquiler[col] = alquiler[col].str.replace('habitación', '', regex=False)
    alquiler[col] = alquiler[col].str.replace('€/mes', '', regex=False)
    alquiler[col] = alquiler[col].str.replace('€/m²', '', regex=False)
    alquiler[col] = alquiler[col].str.replace('€', '', regex=False)


alquiler.head(1)
```

```python
ventas = ventas.drop(['ultima_atualizacion', 'm_construidos', 'm_utiles', 'url', 'referencia', 'ultima_actualizacion'], axis=1)
ventas = ventas.rename(columns={"URL" : "url", "anunciante": "perfil", 'baños': 'banos'})

cols_to_clean = ventas[['habitaciones', 'banos', 'precio', 'precio_metro']]
for col in cols_to_clean:
    ventas[col] = ventas[col].str.replace('baños', '', regex=False)
    ventas[col] = ventas[col].str.replace('baño', '', regex=False)
    ventas[col] = ventas[col].str.replace('habitaciones', '', regex=False)
    ventas[col] = ventas[col].str.replace('habitación', '', regex=False)
    ventas[col] = ventas[col].str.replace('€/mes', '', regex=False)
    ventas[col] = ventas[col].str.replace('€/m²', '', regex=False)
    ventas[col] = ventas[col].str.replace('€', '', regex=False)


ventas.head(1)
```

```python
df = pd.concat([alquiler, ventas], axis=0, ignore_index=True)
df
```

```python
def extraer_y_asignar_caracteristicas(caracteristicas):
    # Convertir la lista en un solo string para facilitar el análisis
    caracteristicas_str = ' '.join(caracteristicas)
    
    # Extraer y asignar cada característica usando expresiones regulares
    
    # Superficie construida
    match = re.search(r'(\d+ m² construidos)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['superficie'] = match.group(0)
    else:
        df['superficie'] = None
    
    # Superficie útil
    match = re.search(r'(\d+ m² útiles)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['superficie_util'] = match.group(0)
    else:
        df['superficie_util'] = None
    
    # Habitaciones
    match = re.search(r'(\d+ habitaciones)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['habitaciones'] = match.group(0)
    else:
        df['habitaciones'] = None
    
    # Baños
    match = re.search(r'(\d+ baños)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['baños'] = match.group(0)
    else:
        df['baños'] = None
    
    # Planta
    match = re.search(r'Planta (\d+)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['planta'] = match.group(1)  # Solo el número de planta
    else:
        df['planta'] = None
    
    # Ascensor
    df['ascensor'] = 'Con ascensor' if re.search(r'Con ascensor', caracteristicas_str, re.IGNORECASE) else None
    
    # Terraza
    df['terraza'] = 'Terraza' if re.search(r'Terraza', caracteristicas_str, re.IGNORECASE) else None
    
    # Garaje
    df['garaje'] = 'Plaza de garaje incluida en el precio' if re.search(r'Plaza de garaje', caracteristicas_str, re.IGNORECASE) else None
    
    # Estado
    df['estado'] = 'Segunda mano/buen estado' if re.search(r'Segunda mano/buen estado', caracteristicas_str, re.IGNORECASE) else None
    
    # Armarios
    df['armarios'] = 'Armarios empotrados' if re.search(r'Armarios empotrados', caracteristicas_str, re.IGNORECASE) else None
    
    # Orientación
    match = re.search(r'Orientación ([^,]+, [^,]+)', caracteristicas_str, re.IGNORECASE)
    if match:
        df['orientacion'] = match.group(1)
    else:
        df['orientacion'] = None
    
    # Amueblado
    df['amueblado'] = 'Amueblado y cocina equipada' if re.search(r'Amueblado y cocina equipada', caracteristicas_str, re.IGNORECASE) else None
    
    # Calefacción
    df['calefaccion'] = 'Calefacción central' if re.search(r'Calefacción central', caracteristicas_str, re.IGNORECASE) else None
    
    # Trastero
    df['trastero'] = 'Trastero' if re.search(r'Trastero', caracteristicas_str, re.IGNORECASE) else None
    


df.apply(lambda row: extraer_y_asignar_caracteristicas(row['caracteristicas']), axis=1)


df
```

```python
patterns = {
    'superficie': re.compile(r'(\d+)\s*m² construidos', re.IGNORECASE),
    'superficie_util': re.compile(r'(\d+)\s*m² útiles', re.IGNORECASE),
    'habitaciones': re.compile(r'(\d+)\s*habitaciones?', re.IGNORECASE),
    'baños': re.compile(r'(\d+)\s*baños?', re.IGNORECASE),
    'terraza': re.compile(r'(Terraza|Balcón)', re.IGNORECASE),
    'garaje': re.compile(r'(Plaza de garaje)', re.IGNORECASE),
    'estado': re.compile(r'(Segunda mano\b.*buen estado)', re.IGNORECASE),
    'armarios': re.compile(r'(Armarios empotrados)', re.IGNORECASE),
    'trastero': re.compile(r'(Trastero)', re.IGNORECASE),
    'orientacion': re.compile(r'Orientación\s([\w\s,]+)', re.IGNORECASE),
    'amueblado': re.compile(r'(Amueblado y cocina equipada|Cocina sin equipar y casa sin amueblar)', re.IGNORECASE),
    'calefaccion': re.compile(r'Calefacción (central|individual|No disponible calefaccion)', re.IGNORECASE),
    'planta': re.compile(r'Planta\s(\d+)(?:ª|\s)?(exterior|interior)', re.IGNORECASE),
    'ascensor': re.compile(r'(Con ascensor|Sin ascensor)', re.IGNORECASE),
    'construccion': re.compile(r'Construido en (\d{4})', re.IGNORECASE),
    'movilidad_reducida': re.compile(r'(Solo acceso exterior adaptado para personas con movilidad reducida)', re.IGNORECASE)
}

def extract_data(row, patterns):
    # Inicializa el diccionario con valores nulos para todas las columnas
    data = {col: np.nan for col in patterns.keys()}
    data['exterior_interior'] = np.nan  # Añadir columna específica para exterior/interior
    if isinstance(row, str):
        row = [row]
    for item in row:
        for key, pattern in patterns.items():
            match = pattern.search(item)
            if match:
                if key == 'planta':
                    data['planta'] = match.group(1)
                    data['exterior_interior'] = match.group(2)
                elif key == 'ascensor':
                    data[key] = 'Sí' if 'Con' in match.group(0) else 'No'
                elif key == 'amueblado':
                    data[key] = 'Amueblado' if 'Amueblado' in match.group(0) else 'No amueblado'
                elif key == 'calefaccion':
                    calef = match.group(1)
                    data[key] = calef if calef != 'No disponible calefaccion' else 'No disponible'
                elif key == 'orientacion':
                    data[key] = match.group(1)
                else:
                    data[key] = match.group(1)
                continue
    return data

# Aplicar la función a la columna 'caracteristicas'
df_extracted = df['caracteristicas'].apply(lambda row: extract_data(row, patterns))

# Convertir la serie de diccionarios a un DataFrame
df_extracted = pd.DataFrame(list(df_extracted))
# Actualizar las columnas de df con los valores extraídos
for col in df_extracted.columns:
    df[col] = df_extracted[col]

df
```
